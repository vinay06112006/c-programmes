```python
import numpy as np
import copy
```

```python
x_train = np.array([[800,2,15],[1000,2,10],[1200,3,8],[1500,3,5],[1800,4,4],[2000,4,3]])
y_train = np.array([[45],[60],[72],[85],[110],[130]])
x_test = np.array([[2200,4,2],[2500,5,1],[2700,5,2],[3000,6,1]])
y_test = np.array([[150],[170],[185],[210]])
print(f"x_train.shape = {x_train.shape}")
print(f"y_train.shape = {y_train.shape}")
```

```python
w_init = np.zeros(x_train.shape[1])
b_init = 0
w_init
```

```python

def cost(x,y,w,b):
    m = x.shape[0]
    cost = 0
    for i in range(m):
        f_wb_i = np.dot(x[i],w) + b
        cost += (f_wb_i - y[i])**2
    cost = cost/(2*m)
    return cost[0]
cost(x_train,y_train,w_init,b_init)
```

```python
def compute_gradients(x,y,w,b) :
   m = x.shape[0] # Number of training examples
   n = x.shape[1] # Number of features

   dj_dw = np.zeros(n) # Initialize dj_dw as a vector of zeros with size n
   dj_db = 0.

   for i in range(m):
    f_wb_i = np.dot(x[i],w) + b
    error = f_wb_i - y[i][0] # Access the scalar value of y[i]
    for j in range(n):
      dj_dw[j] += error * x[i][j]
    dj_db += error

   dj_dw = dj_dw / m
   dj_db = dj_db / m
   return dj_dw,dj_db
```

```python
def gradient_descent(x,y,w,b,alpha,iter,compute_gradients,cost):
  J_history = []
  w = copy.deepcopy(w_init)
  b = b_init
  for i in range(iter):
    dj_dw,dj_db = compute_gradients(x,y,w,b)
    w = w - alpha*dj_dw
    b = b - alpha*dj_db
    '''if i<100000:
      J_history.append(cost(x,y,w,b))
    if i<10 or i%100 == 0:
      print(f"Iteration {i:4}: Cost {J_history[-1]}")'''
  return w,b,J_history
```

```python
w_final,b_final,J_history = gradient_descent(x_train,y_train,w_init,b_init,1.0e-7,10000,compute_gradients,cost)
print(f"w_final = {w_final}")
print(f"b_final = {b_final}")
```

```python
for i in range(x_test.shape[0]):
  print(f"predicted value = {np.dot(x_test[i],w_final) + b_final}")
  print(f"actual value = {y_test[i][0]}")
```
